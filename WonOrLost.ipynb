```python
def extract_scores(raw_text):
  """Extracts lines containing scores from the raw text.

  Args:
    raw_text: The raw text input.

  Returns:
    A list of lines containing scores.
  """

  lines = raw_text.splitlines()
  score_lines = []

  for line in lines:
    if "," in line:
      score_lines.append(line)

  return score_lines

def compute_scores(score_lines):
  """Computes the scores from the given lines.

  Args:
    score_lines: A list of lines containing scores.

  Returns:
    A list of computed scores, where each score is a tuple of the form (player1_score, player2_score).
  """

  scores = []

  for line in score_lines:
    score_parts = line.split(", ")
    player1_score = score_parts[0]
    player2_score = score_parts[-1]
    scores.append((player1_score, player2_score))

  return scores



score_lines = extract_scores(raw_text)
computed_scores = compute_scores(score_lines)

print(score_lines)
print (len(score_lines))
# print(computed_scores)
```

```python
def check_last_element_contains_bp(string):
  """Checks if the last element in a comma-separated string contains "BP".

  Args:
    string: The input string.

  Returns:
    True if the last element contains "BP", False otherwise.
  """

  parts = string.split(",")
  last_part = parts[-1]
  return "BP" in last_part
```

```python
def find_bp_items(score_lines):
  """Finds score_lines in the list that contain "BP".

  Args:
    score_lines: A list of strings.

  Returns:
    A list of score_lines that contain "BP".
  """

  bp_items = []
  for line in score_lines:
    if check_last_element_contains_bp(line):
      bp_items.append(line)
  return bp_items
```

```python
def find_not_bp_items(score_lines):
  """Finds score_lines in the list that contain "BP".

  Args:
    score_lines: A list of strings.

  Returns:
    A list of score_lines that contain "BP".
  """

  not_bp_items = []
  for line in score_lines:
    if not check_last_element_contains_bp(line):
      not_bp_items.append(line)
  return not_bp_items
```
